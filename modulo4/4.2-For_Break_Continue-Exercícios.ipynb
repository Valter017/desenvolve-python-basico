### Q2.

Dado um número inteiro positivo informado pelo usuário, crie um programa em Python que utilize o comando for para calcular e exibir a soma dos números de 1 até o número informado. Ou seja, $1 + 2 + 3 + ... + n$

Exemplo de interação:
```
Digite um número: 5
A soma dos números de 1 a 5 é 15
```


````python
## Escreva e execute seu código aqui

numero = int(input("Digite um número inteiro positivo: "))

soma = 0

for i in range(1, numero + 1):
    soma += i

print(f"A soma dos números de 1 até {numero} é {soma}")

````

### Q3.

Escreva um programa que lê 10 valores inteiros positivos do usuário e ao final imprime a média dos valores digitados com duas casas decimais.

```
Digite 10 números positivos:
3
2
12
14
20
42
19
0
8
1
A média dos valores digitados é 12.10
```

````python
## Escreva e execute seu código aqui
def main():
  
    soma = 0

    for i in range(10):
        while True:
            try:
                valor = int(input(f"Digite o {i+1}º valor inteiro positivo: "))
                if valor <= 0:
                    raise ValueError("O valor deve ser um inteiro positivo.")
                break
            except ValueError as e:
                print(e)
       
        soma += valor

    media = soma / 10

    print(f"A média dos valores digitados é: {media:.2f}")

if __name__ == "__main__":
    main()

````

### Q4.

Você vai criar um sistema que registra os resultados dos jogos do Atlético MG ao longo de um campeonato. Seu sistema vai receber os resultados de todos os jogos do Galo, e deve calcular a pontuação do time sabendo que vitórias valem 3 pontos, empates 1 ponto e derrotas 0 pontos.

Entrada: <br>
A primeira linha de entrada é um inteiro N com a quantidade jogos do galo. Para cada jogo você deve ler 2 inteiros, o primeiro com a quantidade de gols do galo e o segundo com a quantidade de gols do time oponente. 

Saída: <br>
Apresente a soma de vitórias, empates e derrotas do galo, junto com o cálculo da pontuação total.

Ex:
```
Entrada:
4
2
0
1
1
0
3
1
0

Saída:
Vitórias: 2
Empates: 1
Derrotas: 1
Pontuação: 7

```




````python
## Escreva e execute seu código aqui
N = int(input())

vitorias = 0
empates = 0
derrotas = 0
pontuacao_total = 0

for _ in range(N):
    gols_atletico = int(input())
    gols_oponente = int(input())
    
    if gols_atletico > gols_oponente:
        vitorias += 1
        pontuacao_total += 3
    elif gols_atletico == gols_oponente:
        empates += 1
        pontuacao_total += 1
    else:
        derrotas += 1

# Impressão dos resultados
print(f"Vitórias: {vitorias}")
print(f"Empates: {empates}")
print(f"Derrotas: {derrotas}")
print(f"Pontuação total: {pontuacao_total}")
````

### Q6.

Escreva um programa que lê uma quantidade indefinida de valores e informa o maior e o menor valor digitados. A leitura é encerrada quando o usuário digitar o valor 0 (zero).

```
Entrada:
12
-1
29
11
-5
0

Saída:
Maior: 29
Menor: -5
```

````python
## Escreva e execute seu código aqui
maior = float('-inf')  
menor = float('inf')   

while True:
      valor = float(input("Digite um número (ou 0 para sair): "))
   
    if valor == 0:
        break

    if valor > maior:
        maior = valor
    if valor < menor:
        menor = valor

print(f"O maior valor digitado foi: {maior}")
print(f"O menor valor digitado foi: {menor}")

````

### Q7.

Escreva um programa que lê uma quantidade indefinida de valores e informa o **produto** dos valores positivos digitados. A leitura é encerrada quando o usuário digitar o valor 0 (zero).

```
Entrada:
2
-1
1
3
-5
0

Saída:
Produto: 6
```

````python
## Escreva e execute seu código aqui
produto = 1

while True:
    valor = float(input("Digite um valor (0 para sair): "))
    
    if valor == 0:
        break
    
    if valor > 0:
        produto *= valor

print(f"O produto dos valores positivos digitados é: {produto}")

````

### Q8.

Vamos fazer uma calculadora que aceita expressões aritméticas de qualquer tamanho até que o usuário digite a palavra "Fim". Sua calculadora executa apenas as operações "+", "-".

```
Entrada:
5
+
8
-
3
+
2
Fim

Saída:
12
```

````python
## Escreva e execute seu código aqui
def calcular(expressao):
    operadores = {'+': lambda x, y: x + y, '-': lambda x, y: x - y}
    partes = expressao.split()
    resultado = int(partes[0])
    
    for i in range(1, len(partes), 2):
        operador = partes[i]
        numero = int(partes[i + 1])
        resultado = operadores[operador](resultado, numero)
        
    return resultado

def main():
    print("Calculadora simples de expressões (+ ou -)")
    print("Digite expressões para calcular ou 'Fim' para encerrar.")
    
    while True:
        expressao = input("Digite uma expressão (ou 'Fim' para sair): ")
        
        if expressao.lower() == 'fim':
            print("Encerrando a calculadora...")
            break
        
        resultado = calcular(expressao)
        print(f"Resultado da expressão '{expressao}': {resultado}")
    
    print("Calculadora encerrada.")

if __name__ == "__main__":
    main()

````
